### Methods: Plan

* To answer our predictive research question: given the features of age, sex, chest pain types, serum cholesterol, and resting blood pressure of patients as predictors, can we predict the presence or absence of heart disease in patients? We will use the model selection method to select the best model between Ridge, LASSO and Ordinary regression by using cross-validation, which allow us to fit a logistic regression model while addressing multicollinearity and identifying the most relevant predictors, then use the model to make the prediction. This method is useful for predictive modelling and identify the classification of new observations. And logistic regression is appropriate for predicting a binary outcome response variable, giving the outcome as either 1 or 0. Moreover, it is more appropriate to use LASSO or Ridge logistic regression compare to a standard logistic regression model, because these regularization models are better to handle multicollinearity, potential overfitting issues. Therefore, these regularized approaches will help to improve the model’s predictive performance.

*  To apply this LASSO or Ridge logistic regression method, there are several assumptions are required, which include the linearity of relationships, since the LASSO and Ridge logistic regression are linear models, so we will assume there is a linear relationship between predictors and explanatory variables. And we will assume the residuals and errors are independent to avoid the biased estimates of coefficients and standard errors. Moreover, we will also use large sample size, so we combined two datasets to create a large sample, since these models are more stable when the size n is relatively large. Additionally, we will assume the residuals are normally distributed and non-zero variance in predictors. If a predictor have zero variance, then it provides no information to the model. And we will use cross-validation to identify the optimal value for 𝜆.

* Although LASSO or Ridge logistic regression method are used for predicting binary outcome variable, there are some potential limitations and weaknesses of this method, including less effective in terms of model interpretation, while LASSO performs explanatory variables selection by shrinking some coefficients to zero, it may exclude some other relevant predictors that correlated to others, making more challenge to interpret. And while Ridge penalizes coefficients, but it does not perform variables selections, so it includes all predictors, that may lead to a less interpretable model. Moreover,since these regularization models introduces bias to reduce variance, it may cause underfitting problems with poor performance on training and test sets due to simplistic relationships. Additionally, it has limitations with non-linearity relationship between predictors and binary outcome variables, since these methods assume a linear relationship between explanatory variables and response variable. And it also has some limitations in high-dimensional data, when variables are highly correlated, LASSO may have difficulty selecting the right subset of predictors and may exclude correlated variables. On the other hand, Ridge retains all predictors, including irrelevant ones, which reduces its ability to produce sparse models.

* We will perform model selection to select the best model between Ridge and LASSO regressions to fit a logistic regression, and the process is invovling split the training and test sets and cross-validation, which allow us to fit a logistic regression model while addressing multicollinearity and identifying the most relevant predictors, then use the best model to make the prediction. Finally we will use the ROC plot to analyze the prediction accuracy of the results in order to answer our research question.

```r
# Main developer: Selina
# Contributor: Donald
# CV-AUC for a Logistic Regression (without penalization)
set.seed(2024012)

num.folds <- 10

folds <- createFolds(heart_disease_train$num, k=num.folds)

regr.cv <- NULL
for (fold in 1:num.folds) {
train.idx <- setdiff(1:nrow(heart_disease_train), folds[[fold]])
regr.cv[[fold]] <- glm(num ~ ., data=heart_disease_train, subset=train.idx,
                       family="binomial")
    }

pred.cv <- NULL
auc.cv <- numeric(num.folds) 

for (fold in 1:num.folds) {
test.idx <- folds[[fold]]
pred.cv[[fold]] <- data.frame(obs=heart_disease_train$num[test.idx],
pred=predict(regr.cv[[fold]], newdata=heart_disease_train, type="response")[test.idx])
auc.cv[fold] <- roc(obs ~ pred, data=pred.cv[[fold]])$auc
    }

heart_disease_cv_ordinary <- round(mean(auc.cv),7)

cat("Cross-validation AUC for the ordinary logistic model:", 
heart_disease_cv_ordinary)
```

#### Forward Selected Nested In 10-fold Cross-validation

```r
# Main developer: Donald
# Declare prediction function for regsubset class model
predict_model <- function(new_data, coeffs) {
    # Create the model matrix for the new data
    model_matrix <- model.matrix(~ ., data = new_data)  # This creates the design matrix for the data
  
    # Subset the model matrix to only include the selected predictors (same as in the best model)
    selected_predictors <- names(coeffs)
    model_matrix_subset <- model_matrix[, selected_predictors, drop = FALSE]
    
    # Calculate logits (linear combination of features and coefficients)
    logits <- as.vector(model_matrix_subset %*% coeffs)
    
    # Convert logits to probabilities using the logistic function
    probs <- exp(logits) / (1 + exp(logits))
    
    # Return the probabilities (predicted probabilities)
    return(probs)
}
```

```r
# Main developer: Donald
# Contributor: Ankur
# Define 10-Fold Cross-Validation Function with ROC-AUC
cv_regsubsets_auc <- function(data, formula, selection) {
    # Setting the seed
    set.seed(1)
    
    # Create 10 folds
    folds <- sample(1:10, nrow(data), replace = TRUE)
    cv_results <- matrix(NA, nrow = 10, ncol = ncol(data) - 1)
    
    # Loop over each fold
    for (i in 1:10) {
        # Split data into training and testing sets
        train_data <- data[folds != i, ]
        test_data <- data[folds == i, ]
        
        # Fit regsubsets on training data
        reg_fit <- regsubsets(
          x = formula,
          data = train_data,
          method = selection,
          nvmax = 18
        )
        
        # Get the test data predictors and response
        test_matrix <- model.matrix(formula, test_data)
        test_y <- test_data[[all.vars(formula)[1]]] |>
            as.character() |>
            as.numeric()
        valid_rows <- rownames(test_matrix)
        test_y <- test_y[rownames(test_data) %in% valid_rows]
        
        # Evaluate the model for each subset size
        for (j in 1:(ncol(data) - 1)) {
            
            # Get coefficients for the jth model
            coeffs <- coef(reg_fit, id = j)
    
            valid_rows <- complete.cases(test_matrix[, names(coeffs), drop = FALSE])
            test_y <- test_y[valid_rows]

            # Predict the target value in test_data with the new model
            prediction <- predict_model(new_data = test_data, coeffs = coeffs)
            
            # Calculate ROC-AUC
            roc_obj <- roc(
                response = test_y, 
                predictor = prediction, 
                levels = c(0, 1), 
                direction = "<")
            
            # Store the AUC value
            cv_results[i, j] <- auc(roc_obj)  
        }
    }
  
    # Return average ROC-AUC for each model size
    return(colMeans(cv_results))
}
```

```r
# Main developer: Donald
# Perform forward selection to the training data
cv_error_auc <- cv_regsubsets_auc(heart_disease_train, num ~ ., "forward")

optimal_size <- which.max(cv_error_auc)
cat("Optimal model size:", optimal_size, "\n")
cat("Optimal ROC-AUC:", max(cv_error_auc), "\n")
```

## Discussion

```r
as.data.frame(as.matrix(coef(heart_disease_ridge_max_AUC)))
```

From our analysis, the Ridge model we selected achieved a relatively high AUC score of 0.903 on the test set, indicating it is very proficient at distinguishing between the positive and negative classes—in our context, a patient with heart disease (positive class) and a patient without heart disease (negative class). Essentially, this means that if our model is given a pair consisting of one patient with heart disease and one without, it will correctly assign a higher probability of having heart disease to the patient with heart disease about 90.3% of the time.

On the training set, the Ridge model with the minimized lambda had a marginally higher accuracy compared to other models, with an improvement of only 0.003 over the next most accurate model. This finding is somewhat surprising given that Ridge regression does not eliminate any variables; instead, it suggests that all variables played a significant role in predicting heart disease, even if some multicollinearity existed among them.

There are a few ways in which this model could be improved. One area which could be improved is the way we handled data processing. During data processing, we encountered a few columns with several missing values. When creating this model, we decided to remove them to ensure our model worked. However, this removed a large number of observations from our dataset. Although outside the scope of this course, we could try imputing the data by replacing the missing values with some measure of central tendency like the average or median of the column. This could potentially improve the accuracy of the model by giving the model more observations to be trained on.

Another improvement could come from the models we chose. When looking at the average cross-validation scores for our model, we can see that the models with the highest accuracy were both the Ridge models and both the LASSO models. It is important to note that the accuracies of these models were not that far apart, with the range of AUC scores for these models only being  around 0.025. As these models use either a L1 or L2 regularlization method, we could potential see improvements if we used a model that combines both of these methods, such as Elastic Net. 

It is important to note that this study only used data from Cleveland and Hungary. As such, it is possible that this model is primarily effective at predicting heart disease in patients from these specific regions, potentially due to unique biological traits of patients in these areas. Therefore, future studies should consider developing models using data from a variety of regions and investigating whether there are significant differences between regions. This analysis could potentially be conducted using an ANOVA test, treating each region as its own group.
